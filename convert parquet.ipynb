```python
import pandas as pd

# Load Excel
df = pd.read_excel("MASTER PLU PRINSIPAL.xlsx")

# 1. Bersihkan nama kolom
df.columns = df.columns.astype(str).str.strip()

# 2. Hapus kolom 'nan'
df = df.loc[:, df.columns != 'nan']

# 3. Drop duplicate column
df = df.loc[:, ~df.columns.duplicated()]

# 4. Fix semua kolom object → string
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.strip()

# 5. Save ke parquet
df.to_parquet("master_plu.parquet", index=False)

print("✅ Convert berhasil tanpa error!")
```
